```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Diveorid_survey_results.csv')
df = df.fillna(0)
df.columns = ['time','shipping','device','how','why','phone','suggestion','innovation','price','computer','convinient','div','cam','log','aha','color']
df['phoneCo'] = df.phone.map(lambda x: str(x).lower()).map( lambda x:
    'apple' if 'iphone' in x or 'apple' in x
    else 'samsung' if 'sam' in x or 'note' in x or 'gal' in x or 's10' in x or 's9' in x
    else 'motorola' if 'moto' in x
    else 'one' if 'one' in x
    else 'huawei' if 'hua' in x or 'mate' in x or 'honor' in x
    else 'pixel' if 'pix' in x
    else 'lg' if 'lg' in x
    else 'ruggear' if 'rug' in x
    else 'both' if 'android' in x
    else 'Unknown' if '0' == x
    else x)
df['how'] = df.how.map(lambda x: str(x).lower()).map( lambda x:
    'FBinsta' if 'face' in x 
    else 'KSbrowse' if 'browsing' in x 
    else 'KSupdate' if 'update' in x
    else 'MediaBlog' if 'media' in x
    else 'Google' if 'google' in x 
    else x)
df.head(3)
```

```python
def print_per(df):
    for i in range(len(df.phoneCo)):
        try:
            print(df.phoneCo.value_counts().index[i]+': ' + str(round(df.phoneCo.value_counts()[i]/len(df)*100,2))+'%')
        except:
            pass
    print('total: ' +str(len(df)))
def plot_print(df,str):
    plt.figure(figsize=(10,3))
    df.phoneCo.value_counts().plot(kind='bar')
    plt.title(str+' x phone')
    print_per(df)
def pp(str):
    df2 = df[df[str] == 1]
    plot_print(df2,str)

def print_per2(df):
    for i in range(len(df.how)):
        try:
            print(df.how.value_counts().index[i]+': ' + str(round(df.how.value_counts()[i]/len(df)*100,2))+'%')
        except:
            pass
    print('total: ' +str(len(df)))
def plot_print2(df,str):
    plt.figure(figsize=(10,3))
    df.how.value_counts().plot(kind='bar')
    plt.title(str+' x how')
    print_per2(df)
def hp(str):
    df2 = df[df[str] == 1]
    plot_print2(df2,str)

def print_per3(df):
    for i in range(len(df.shipping)):
        try:
            print(df.shipping.value_counts().index[i]+': ' + str(round(df.shipping.value_counts()[i]/len(df)*100,2))+'%')
        except:
            pass
    print('total: ' +str(len(df)))
def plot_print3(df,str):
    plt.figure(figsize=(10,3))
    df.shipping.value_counts().plot(kind='bar')
    plt.title(str+' x shipping')
    print_per3(df)
def sp(str):
    df2 = df[df[str] == 1]
    plot_print3(df2,str)
```
